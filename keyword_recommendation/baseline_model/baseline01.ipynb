```python
import pandas as pd
import pandasql as psql
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec
```

```python
# 조회수 테이블 불러오기

view = pd.read_csv('view_result.csv')
view.head()
```

```python
# 공감수 테이블 불러오기

empathy = pd.read_csv('empathy_result.csv')
empathy.head()
```

```python
# SQL을 활용해서 두 개의 테이블 합치기

query = 'SELECT v.title, v.views, IFNULL(e.empathies, 0) AS empathies, v.date ' \
'FROM view v LEFT JOIN empathy e ON v.title = e.title'
result = psql.sqldf(query, locals())
result
```

```python
result.to_csv('result.csv', encoding='utf-8-sig')
```

```python
# 수동 크롤링으로 게시물 내용 긁어와서 새로운 열로 추가

top_result = pd.read_csv('result_text.csv', encoding='euc-kr').drop('Column1', axis=1)
top_result
```

```python
# text에서 명사 추출

okt = Okt()
top_result['nouns'] = top_result['text'].apply(lambda x: [w for w, t in okt.pos(str(x)) if t == 'Noun'])
top_result
```

```python
# 추출한 명사의 TF-IDF 계산

corpus = [' '.join(tokens) for tokens in top_result['nouns']]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(corpus)
tfidf_scores = tfidf_matrix.mean(axis=0).A1
keywords = pd.DataFrame({
    'keyword' : vectorizer.get_feature_names_out(),
    'score' : tfidf_scores
}).sort_values('score', ascending=False).head(30)

print('인기 키워드 Top 30: ')
keywords
```

```python
# word2vec 모델 학습

model = Word2Vec(sentences=top_result['nouns'], 
                 vector_size=100, 
                 window=5, 
                 min_count=2, 
                 sg=1)
```

```python
# TF-IDF 상위 30개 키워드별 유사 단어 top 5 추출

top_keywords = keywords['keyword'].head(30).tolist()
recommendations = []
for kw in top_keywords:
    try:
        similar_words = [w for w, sim in model.wv.most_similar(kw, topn=8)]
        recommendations.append({
            '핵심키워드' : kw,
            '연관키워드' : ', '.join(similar_words),
        })
    except KeyError:
        continue

recommend_df = pd.DataFrame(recommendations)
recommend_df
```
